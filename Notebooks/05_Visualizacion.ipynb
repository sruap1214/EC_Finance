### Interfaz `pyplot`

La interfaz `pyplot` proporciona una serie de funciones que operan sobre un estado global.


- No especificamos sobre qué gráfica o ejes estamos actuando. 


- Es una forma rápida y cómoda de crear gráficas pero perdemos parte del control.


- El paquete pyplot se suele importar bajo el alias *plt*, de modo que todas las funciones se acceden a través de `plt.<funcion>`. 


- La función más básica es la función plot:

```python
x = [0, 0.1, 0.2, 0.5]
y = [1, -1, 0.5, 2]
plt.plot(x, y)
```

```python
# Carga los módulos necesarios
import numpy as np
import matplotlib.pyplot as plt
# Crea el arreglo x de cero a cien con cien puntos
x = np.linspace(0, 10, 100)
# Crea el arreglo y con la función indicada
ys = np.sin(x)
yc = np.cos(x)
# Crea una figura
plt.figure(2)
# imprime la figura
plt.plot(x,ys,'r',x,yc,'-g')
plt.grid(True)
```

Como hago para configurar para que se grafiquen a parte?

```python
import numpy as sp
import matplotlib.pyplot as plt
# Crea el arreglo x de cero a cien con cien puntos
x = sp.linspace(0, 10, 100)
# Crea el arreglo y con la función indicada
ys = sp.sin(x)
yc = sp.cos(x)

# Crea figura de tamaño 6,6 inches
fig = plt.figure(figsize=(6, 6))
ax1 = fig.add_subplot(121)     # Subplot de 1 fila, 2 columnas, Primera columna
plt.plot(x,ys,"r", label = "sin(x)")
plt.legend()

ax2 = fig.add_subplot(122)     # Subplot de 1 fila, 2 columnas, Segunda columna
plt.plot(x,yc,"g", label = "cos(x)")
plt.legend()

fig.savefig('img/seno.png')
```

La función `plot` recibe una sola lista (si queremos especificar los valores y) o dos listas (si especificamos $x$ e $y$). 


- Naturalmente si especificamos dos listas ambas tienen que tener la misma longitud.


La tarea más habitual a la hora de trabajar con `matplotlib` es representar una función. 


- Lo que tendremos que hacer es definir un dominio y evaluarla en dicho dominio. Por ejemplo:


$$f(x)=e^{-x^2}$$

```python
# defino la función que quiero graficar
def f(x):
    return np.exp(-x ** 2)

x = np.linspace(-1, 5, num=100)

plt.plot(x, f(x), label="Función f(x)")
plt.xlabel("Eje $x$")
plt.ylabel("$f(x)$")
plt.legend()
plt.title("Función $f(x)=e^{-x^2}$")
```

### Personalización

- La función `plot` acepta una serie de argumentos para personalizar el aspecto de la función. 


- Con una letra podemos especificar el color, y con un símbolo el tipo de línea.

```python
plt.plot(x, f(x), 'g-')
plt.plot(x, 1 - f(x), 'b--')
```

Esto en realidad son códigos abreviados, que se corresponden con argumentos de la función `plot`:

```python
plt.plot(x, f(x), color='black', linestyle='', marker='o')
plt.plot(x, 1 - f(x), c='g', ls='--')
```

La lista de posibles argumentos y abreviaturas está disponible en la documentación de la función `plot` la podrás encontrar en este [enlace](https://matplotlib.org/2.0.2/api/pyplot_api.html).

### Otro tipo de gráficas: Scatter y Contour

La función `scatter` muestra una nube de puntos, con posibilidad de variar también el tamaño y el color.

```python
N = 100
x = np.random.randn(N)
y = np.random.randn(N)

plt.scatter(x, y)
```

Con `s` y `c` podemos modificar el tamaño y el color respectivamente. 

- Para el color, a cada valor numérico se le asigna un color a través de un mapa de colores; ese mapa se puede cambiar con el argumento `cmap`. Esa correspondencia se puede visualizar llamando a la función `colorbar`.

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
plt.colorbar();  # show color scale
```

La función `contour` se utiliza para visualizar las curvas de nivel de funciones de dos variables y está muy ligada a la función `np.meshgrid`. Veamos un ejemplo:

$$f(x,y) = cos(x) + sin^2(y)$$

```python
def f(x, y):
    return np.cos(x) + np.sin(y) ** 2

x = np.linspace(-2, 2)
y = np.linspace(-2, 2)
xx, yy = np.meshgrid(x, y)

plt.contour(xx, yy, f(xx, yy))
plt.colorbar()
```

La función `contourf` es casi idéntica pero rellena el espacio entre niveles. 

- Podemos especificar manualmente estos niveles usando el cuarto argumento:

```python
zz = f(xx, yy)
plt.contourf(xx, yy, zz, np.linspace(-0.5, 2.0),cmap=plt.cm.rainbow)
plt.colorbar()
```

## Visualización de errores

La visualización de errores es importante en muchas ocasiones mas que el valor exacto que se presenta como resultado final. En muchas ocasiones estos errores se presentan como una desviación estandar del valor obtenido.

## Subplots, textos y anotaciones

En ocasiones es importante realizar varias gráficas en figuras diferentes para mostrar como son los datos a representar. Estas gráficas pueden ser una dentro de otra, una grilla, o plantillas mas complejas. La primera forma para lograr estas subgráficas es utilizando el método `plt.subplot`, asi

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=0.4)
for i in range(1, 7):
    plt.subplot(2, 3, i)
    plt.text(0.5, 0.5, str((2, 3, i)),
             fontsize=18, ha='center')
```

La primera línea de código crea la figura completa y almacena el objeto creado en `fig`. Posteriormente se indica como es la relación de aspecto entre los diferentes subplot con el método `subplots_adjust`. Los dos primeros argumentos de subplot hace referencia a la dimensión de la grilla, en este caso de 2 filas por 3 columnas. El último parámetro, hace referencia a la figura actual que se va a escribir. 

La segunda forma para incluir subfiguras, es con el método `plt.axes`, el cual por defecto crea un objeto de eje que llena la figura completa. Este método adicionalmente recibe un argumento adicional con cuatro números que indicar el sistema coordenado de este nuevo eje asi `[left, bottom, width, height]`. 

```python
ax1 = plt.axes()  # standard axes
ax2 = plt.axes([0.65, 0.65, 0.2, 0.2])
```

En ocasiones es importante hacer alguna anotación en la gráfica con el objetivo de marcar algun punto relevante que se quiere informar. Para lograr lo anterior, es posible agregar a los ejes creado anotaciones utilizando el método `ax.text`.

**Ejemplo:** tenemos los datos de la tasa de nacimientos que ocurren en Estados unidos, y quisieramos graficar su comportamiento en el tiempo. Adicionalmente, se solicita que se anote los días feriados para entender que pasa en dichas fechas. 

```python
import numpy as np
import pandas as pd
births = pd.read_csv('data/births.csv')
births.head(10)
```

```python
births.drop(births[births['day'] > 28].index,inplace=True)
birth_mean = births.groupby(['month','day'])['births'].mean().reset_index()
birth_mean['year'] = 2012
birth_mean.head()
```

```python
birth_mean['DateTime'] = pd.to_datetime(birth_mean[['year','month','day']],format='%Y%m%d')
birth_mean.head()
```

```python
fig, ax = plt.subplots(figsize=(12, 4))
births_by_date = birth_mean[['DateTime','births']]
births_by_date = births_by_date.set_index('DateTime')
births_by_date.plot(ax=ax);

# Add labels to the plot
style = dict(size=10, color='grey')

ax.text('2012-1-1', 3950, "New Year's Day", **style)
ax.text('2012-7-4', 4250, "Independence Day", ha='center', **style)
ax.text('2012-9-4', 4850, "Labor Day", ha='center', **style)
ax.text('2012-10-31', 4600, "Halloween", ha='right', **style)
ax.text('2012-11-25', 4450, "Thanksgiving", ha='center', **style)
ax.text('2012-12-25', 3850, "Christmas ", ha='right', **style)

# Label the axes
ax.set(title='USA promedio de nacimientos por año (1969-1988)',
       ylabel='valor promedio')
```

## Seaborn

Matplotlib es una herramienta muy útil y poderosa a la hora de realizar gráficas, sin embargo, hay algunas desventajas que tiene:
 - Matplotlib está basado en Matlab, por lo que muchas decisiones no están de acuerdo al estandar de Python
 - La libreria es de bajo nivel, deja realizar cosas y modificaciones a todos los niveles, pero requiere de bastante programación
 - Matplotlib fue creada antes de Pandas, porque lo que no fue pensada en optimizar su uso. 
 
La forma de resolver muchas de estas cosas es utilizando la libreria `Seaborn`. Es una API encima de Matplotlib la cual tiene definiciones por defecto y simplifica su uso. 

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd

# Create some data
rng = np.random.RandomState(0)
x = np.linspace(0, 10, 500)
y = np.cumsum(rng.randn(500, 6), 0)
```

```python
# Plot the data with Matplotlib defaults
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left');
```

Notese que la gráfica contiene toda la información, pero estéticamente puede ser abrumadora. Como se habia explicado, Seaborn puede trabajar encima de Matplotlib y producir su propia salida

```python
import seaborn as sns
sns.set()

# same plotting code as above!
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left');
```

### Explorando los gráficos de Seaborn

La principal idea de Seaborn es generar una interfaz de alto nivel para realizar gráficas y explorar datos de forma rápida. A continuación se explorará algunos dataset propios de la libreria.

#### Histogramas, KDE, y densidades

A menudo es importante graficar histogramas o distribución conjunta de variables. Desde Matplotlib es fácil utilizando el comando `plt.hist`

```python
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])
print(data.head())

for col in 'xy':
    plt.hist(data[col], alpha=0.5)
```

Por ejemplo, en vez de dibujar el histograma, se puede graficar con Searbon una estimación de la curva del mismo

```python
for col in 'xy':
    sns.kdeplot(data[col], shade=True)
```

```python
sns.displot(data, kde=True)
```

```python
sns.displot(data, x="x", y="y", kind="kde")
```

Se podrian graficar ambas distribuciones al mismo tiempo utilizando `jointplot`

```python
sns.jointplot(data = data, x = 'x', y = 'y', 
              kind = 'kde',
              joint_kws = {'color' : 'k'},
              marginal_kws = {'lw' : 4, 'shade' : True})  
```

```python
sns.jointplot(data = data, x = 'x', y = 'y', 
              kind = 'hex')    
```

#### Plot en parejas

#### Histogramas via subconjuntos

A veces es importante realizar gráficas de histogramas o distribuciones dependiendo de condiciones que se quieran comparar. El método `FacetGrid` permite realizar esto de la manera más sencilla. 

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']
grid = sns.FacetGrid(tips, 
                     row="sex",
                     col="time",
                     margin_titles=True)
grid.map(sns.histplot, "tip_pct",
         bins=np.linspace(0, 45, 15));
```

#### Cat plot

Cat plot sirve para realizar gráficas de variables categoricas y mostrar su relación entre ellas. 

```python
g = sns.catplot(x = "day", 
                y = "total_bill", 
                hue = "sex", 
                data = tips, 
                kind="box")
g.set_axis_labels("Day", "Total Bill");
```

```python
planets = sns.load_dataset('planets')
planets.head()
```

Notese que tanto la columna de `split` como `final` pandas las cargo por defecto como una cadena. Utilizaremos la libreria de datetime para convertirla

```python
import datetime

def convert_time(s):
    h, m, s = map(int, s.split(':'))
    return datetime.timedelta(hours=h, minutes=m, seconds=s)

data = pd.read_csv('./data/marathon-data.csv',
                   converters={'split':convert_time, 'final':convert_time})
data.head()
```

```python
data.dtypes
```

```python
(data['split']/np.timedelta64(1000000000, 'ns')).astype(int)
```

```python
data['split_sec'] = (data['split']/np.timedelta64(1000000000, 'ns')).astype(int)
data['final_sec'] = (data['final']/np.timedelta64(1000000000, 'ns')).astype(int)
data.head()
```

```python
g = sns.jointplot(x = "split_sec", 
                  y = "final_sec", 
                  data = data,
                  kind='hex')
g.ax_joint.plot(np.linspace(4000, 16000),
                np.linspace(8000, 32000), ':k')
```

Notese que la línea de la mitad seria alguien que lleve la misma velocidad constante en toda la maraton. Que la distribución este por encima, indica que la mayoria de los participantes disminuyen su velocidad al final de la maraton. Para aquellas persona que hacen alreves, donde son más rápidos al final son conocidos "negative-split"

A continuación creemos una medida de el grado en el cual se realiza el negative split. Donde la diferencia sea menor a cero, entonces la persona realizar este tipo de situación

```python
data['split_frac'] = 1 - 2 * data['split_sec'] / data['final_sec']
data.head()
```

```python
sns.displot(data['split_frac'], kde=False);
plt.axvline(0, color="k", linestyle="--");
```

```python
sum(data.split_frac < 0)
```

Apenas 250 del total de participantes realizan la maraton de esta forma. Analicemos si existe alguna relación entre el split_frac y otras variables 

```python
g = sns.PairGrid(data, vars=['age', 'split_sec', 'final_sec', 'split_frac'],
                 hue='gender', palette='RdBu_r')
g.map(plt.scatter, alpha=0.8)
g.add_legend();
```

Parece que el  `split_fraction` no se correlaciona con la edad, pero si un poco con el tiempo final: los corredores más rápido o que se demoran menos tiempo en la competencia, tienden a tener un `split_fraction` negativo o igual a cero.
